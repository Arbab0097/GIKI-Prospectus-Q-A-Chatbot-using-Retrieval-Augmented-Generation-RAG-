```python
# Install if not already installed
!pip install sentence-transformers langchain chromadb python-dotenv

import os
from dotenv import load_dotenv
from langchain.vectorstores import Chroma
from langchain.embeddings import HuggingFaceEmbeddings
from uuid import uuid4

# Load API key from .env
load_dotenv()
hf_token = os.getenv("HUGGINGFACEHUB_API_TOKEN")

# Initialize Hugging Face embeddings
embeddings_model = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")

# Example documents
chunks = [
    "This is the first document.",
    "This is the second document."
]
uuids = [str(uuid4()) for _ in chunks]

# Initialize Chroma vector store
vector_store = Chroma(embedding_function=embeddings_model)

# Add documents
vector_store.add_documents(documents=chunks, ids=uuids)

print("Documents added successfully using Hugging Face embeddings!")
```
